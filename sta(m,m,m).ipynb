```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv("Employee_2.csv")
```

```python
# 1. Summary statistics (mean, median, min) for numeric variables
numeric_cols = df.select_dtypes(include=[np.number])
summary_stats = numeric_cols.agg(['mean', 'median', 'min'])
print("1. Summary Statistics for Numeric Variables:")
print(summary_stats)
print("\n")
```

```python
#2. Grouped summary statistics by a categorical variable (Education)
grouped_salary_stats = df.groupby('Education')['Salary'].agg(['mean', 'median', 'min'])
print("2. Salary Summary Statistics Grouped by Education:")
print(grouped_salary_stats)
print("\n")
```

```python
# Create a list of salary values for each education group
salary_grouped_list = df.groupby('Education')['Salary'].apply(list)
print("List of Salary Values by Education:")
print(salary_grouped_list)
print("\n")
```

```python
# 3. Count of each JOB_ID (substituting with 'Education' as proxy)
job_id_count = df['Education'].value_counts()
print("3. Count of Each 'JOB_ID' (Education used as substitute):")
print(job_id_count)
print("\n")
```

```python
# 4. Visualization for Salary column
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
