```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Employee_2.csv'  # Update with your full path if needed
df = pd.read_csv(file_path)
```

```python
# 1. Summary Statistics: max, standard deviation, covariance
numeric_cols = df.select_dtypes(include=[np.number])
```

```python
# Max and Std Dev
max_stats = numeric_cols.max()
std_stats = numeric_cols.std()
cov_matrix = numeric_cols.cov()

print("1. Summary Statistics for Numeric Variables:\n")
print("Maximum:\n", max_stats, "\n")
print("Standard Deviation:\n", std_stats, "\n")
print("Covariance Matrix:\n", cov_matrix, "\n")
```

```python
# 2. Grouped summary statistics by a categorical variable (e.g., Education)
grouped_stats = df.groupby('Education')['Salary'].agg(['max', 'std', 'mean'])
print("2. Salary Summary Grouped by Education:\n", grouped_stats, "\n")
```

```python
# Create list of salaries for each category in Education
salary_grouped_list = df.groupby('Education')['Salary'].apply(list)
print("List of Salary Values by Education:\n", salary_grouped_list, "\n")
```

```python
# 3. Group by 'Education' as a proxy for 'JOB_ID' and count
job_id_count = df['Education'].value_counts()
print("3. Count of Each JOB_ID (Education used as substitute):\n", job_id_count, "\n")
```

```python
# 4. Visualization for a column (e.g., 'Age')
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='lightgreen')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
