```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('Salary_Data.csv')
```

```python
# Prepare the data
X = df[['YearsExperience']]
y = df['Salary']
```

```python
# Train the linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict salary for 15 years of experience
experience = 15
predicted_salary = model.predict([[experience]])
print(f"Predicted salary for {experience} years of experience: ₹{predicted_salary[0]:,.2f}")
```

```python
# Plot the data and regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter([experience], predicted_salary, color='green', s=100, label='Prediction (15 yrs)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Experience vs Salary')
plt.legend()
plt.grid(True)
plt.show()
```
